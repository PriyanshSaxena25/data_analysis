```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
```

```python
%matplotlib inline
plt.plot(yield_apples)
```

```python
years = [2010,2011,2012,2013,2014,2015]
```

```python
plt.plot(years,yield_apples);
```

##Axis Labels
plt.xlabel plt.ylabel can be used for x axis and y axis to give them a name

```python
plt.plot(years,yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield(Tons per Hectare');
```

##Plotting Multiple Lines 


```python
oranges = [0.962, .941, 0.930, 0.923, 0.918, 0.908]
```

```python
plt.plot(years,yield_apples)
plt.plot(years,oranges)
plt.xlabel('Year')
plt.ylabel('Yield (Tons Per Hectare)');
```

By plotting 2 different lines on a saingle graph easier compariosn can be made 


##Plot Title and Legends

```python
plt.plot(years,yield_apples)
plt.plot(years,oranges)
plt.xlabel('Year')
plt.ylabel('Yield (Tons Per Hectare)')
plt.title("Crop Yields")
plt.legend(['Apples',"Oranges"]);
```

##Line Markers
Marker is an argument for plot method

```python
plt.plot(years,yield_apples, marker = 'o')
plt.plot(years,oranges, marker = 'x')
plt.xlabel('Year')
plt.ylabel('Yield (Tons Per Hectare)')
plt.title("Crop Yields")
plt.legend(['Apples',"Oranges"]);
```

##FMT Arguement
fmt = '[marker][line][color]'

```python
plt.plot(years,yield_apples, 'o-b')
plt.plot(years,oranges,'x--r')
plt.xlabel('Year')
plt.ylabel('Yield (Tons Per Hectare)')
plt.title("Crop Yields")
plt.legend(['Apples',"Oranges"]);
```

##Extras

##Scatter Plot
The values of 2 variables are plotted as points on a 2D grid

```python
flowers_df = sns.load_dataset("iris") #predefined dataset
flowers_df
```

```python
flowers_df.species.unique() 
```

```python
plt.plot(flowers_df.sepal_length,flowers_df.sepal_width)
# plt.plot(flowers_df.sepal_width,'x-r');
```

```python
sns.scatterplot(flowers_df.sepal_length,flowers_df.sepal_width)
```

##Hue
we can add hue to the scatter plot to visualze it more clearly 
as we can see in the plot there are majorly three clusters forming up. This can be hypthisized as three species forming three clusters 

```python
sns.scatterplot(flowers_df.sepal_length,
                flowers_df.sepal_width,
                hue = flowers_df.species,s=100)
```

##Seabrn an Pandas
Seabirn has great support for pandas library threrefore we can directly pass culmn names ino the plottng function and use data attribute to specify the data frame.

```python
sns.scatterplot('sepal_length',
                'sepal_width',
                hue = 'species',
                data = flowers_df)
```
