```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
```

```python
%matplotlib inline
plt.plot(yield_apples)
```

```python
years = [2010,2011,2012,2013,2014,2015]
```

```python
plt.plot(years,yield_apples);
```

##Axis Labels
plt.xlabel plt.ylabel can be used for x axis and y axis to give them a name

```python
plt.plot(years,yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield(Tons per Hectare');
```

##Plotting Multiple Lines 


```python
oranges = [0.962, .941, 0.930, 0.923, 0.918, 0.908]
```

```python
plt.plot(years,yield_apples)
plt.plot(years,oranges)
plt.xlabel('Year')
plt.ylabel('Yield (Tons Per Hectare)');
```

By plotting 2 different lines on a saingle graph easier compariosn can be made 


##Plot Title and Legends

```python
plt.plot(years,yield_apples)
plt.plot(years,oranges)
plt.xlabel('Year')
plt.ylabel('Yield (Tons Per Hectare)')
plt.title("Crop Yields")
plt.legend(['Apples',"Oranges"]);
```

##Line Markers
Marker is an argument for plot method

```python
plt.plot(years,yield_apples, marker = 'o')
plt.plot(years,oranges, marker = 'x')
plt.xlabel('Year')
plt.ylabel('Yield (Tons Per Hectare)')
plt.title("Crop Yields")
plt.legend(['Apples',"Oranges"]);
```

##FMT Arguement
fmt = '[marker][line][color]'

```python
plt.plot(years,yield_apples, 'o-b')
plt.plot(years,oranges,'x--r')
plt.xlabel('Year')
plt.ylabel('Yield (Tons Per Hectare)')
plt.title("Crop Yields")
plt.legend(['Apples',"Oranges"]);
```

##Extras
Only marked llines can be also plotted by just specifying a marker type.
Aspect ratio and size of the plot can also be modified using figure.(figsize())
