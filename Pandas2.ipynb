##Working with Dates
The covid data which we are working is a time series data and such type of data are genreally associated with date


```python
covid_df.date #dtype is object
```

Here the datatype is retuned as an object which means that pandas does not know right now that the column is a data object
This can be easily fixed by using todate function

```python
covid_df.date #dtype is datetime64
```

 We can now extract different parts of the data into separate columns, using the DatetimeIndex class.

```python
covid_df['year'] = pd.DatetimeIndex(covid_df.date).year
covid_df['month'] = pd.DatetimeIndex(covid_df.date).month
covid_df['day'] = pd.DatetimeIndex(covid_df.date).day
covid_df['weekday'] = pd.DatetimeIndex(covid_df.date).weekday
```

```python
covid_df
```

##Using the above operations we can query the data for a particular month, week or even a particular day

```python
covid_df_may = covid_df[covid_df.month == 5]
```

```python
covid_df_may
```

```python
covid_df_may_cases = covid_df_may.new_cases.sum()
covid_df_may_deaths = covid_df_may.new_deaths.sum()
covid_df_may_tests = covid_df_may.new_tests.sum()
print(covid_df_may_cases)
print(covid_df_may_deaths)
print(covid_df_may_tests)
```

The above process can achieved like this too

```python
covid_df_may_totals = covid_df_may[['new_cases','new_deaths','new_tests']]
covid_df_may_totals #subset of covid_df_may
```

```python
covid_df_may_totals.sum() #same result in much more representable way
```
