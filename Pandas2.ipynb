##Querying and sorting data 
this can be achived using smiple compariosn operators

```python
import pandas as pd
```

```python
import urllib.request
urllib.request.urlretrieve('https://hub.jovian.ml/wp-content/uploads/2020/09/italy-covid-daywise.csv','italy-covid-daywise.csv')
```

```python
covid_df = pd.read_csv('italy-covid-daywise.csv')
```

```python
high_new_cases = covid_df.new_cases > 1000 # returns bool values for records 
```

```python
high_new_cases
```

##Using the boolean series
we can pass this boolean series to the data frame and get output as follows

```python
covid_df[high_new_cases] #only those cases were returned which had boolean value true i.e >1000
```

```python
high_cases_df = covid_df[covid_df.new_cases > 1000]
high_cases_df #same as above 
```

##New Coulmns can be added just like a dictionary logic

```python
covid_df['postive_rate'] = covid_df.new_cases/covid_df.new_tests
```

```python
covid_df
```

```python
covid_df.drop(columns=['postive_rate'] , inplace=True) #dropping/deleting coulmns
```

##Sorting data

```python
covid_df.sort_values('new_cases', ascending=False).head(10) #returns first 10 max new cases
```

```python
covid_df.sort_values('new_deaths', ascending=False).head(10) #returns days with max new deaths 
```

1. Replcae it with 0
2. replace it with average of entire column
3. replace it with average of previous and next records
4. Discard the row entierly

the method that will be choosen by us need a proper context of the data and thus a proper decision should be made
Here, we will be replacing the value by average of above below vlaue

```python
covid_df.at[172,'new_cases'] = (covid_df.at[171,'new_cases'] + covid_df.at[173,'new_cases'])/2
```
